```python
with open("data/0.txt", "r") as f:
	text = f.read()

print(len(text))
print(text[:50])
```

```python
chars = sorted(list(set(text)))
print(chars)

vocab_size = len(chars)
print(vocab_size)
```

```python
stoi = {c:i for i, c in enumerate(chars)}
itos = {i:c for i, c in enumerate(chars)}

encode_string = lambda s: [stoi[c] for c in s]
print(encode_string("hi"))

decode_string = lambda l: "".join([itos[i] for i in l])
print(decode_string([51, 52]))
```
